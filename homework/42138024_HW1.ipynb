如下的代码框中
利用hist()方法做出了连续变量的直方图，根据直方图可以获得每个变量的大致分布情况
Income的分布呈现极为明显的右偏现象，大部分取值集中在数值较小的部分。
Limit，Rating，Cards，Balance的分布为一个偏态的分布，数据分布右偏，存在少量的较大值

```python
import matplotlib.pyplot as plt
credit.hist(figsize=(20,15))
```

```python
import seaborn as sns
sns.set_theme(rc={'figure.figsize':(12,8)})
sns.set_style('ticks')
```

如下的代码框绘制出Gender箱线图
从图中可知，男性以及女性的Balance的平均值和最小值相同，但女性的Balance的25%分位数和75%分位数高于男性
但是男性的Balance的最大值高于女性

```python
sns.boxplot(x='Gender', y='Balance', data=credit)
```

如下的代码框绘制出Students的箱线图
从图中可以看出，学生群体的Balance的平均值，25%分位数，75%分位数明显高于非学生群体，
学生群体中Balance的最大值小于非学生群体，但是非学生群体中的Balance数据存在异常值

```python
sns.boxplot(x='Student', y='Balance', data=credit)
```

```python
sns.boxplot(x='Married', y='Balance', data=credit)
```

如下的代码框中绘制出Ethnicity的箱线图
从图中可以看出，Caucasian和African American的Balance均值以及25%分位数，75%分位数接近；Asian的Balance均值略低于二者，
25%分位数比二者低，75%分位数比两者高。Asian和African American的最大值接近，Caucasian的最大值则低于这两类人群，且存在
异常值

```python
sns.boxplot(x='Ethnicity',y='Balance',data=credit)
```

如下的代码框绘制出Education的箱线图
如图所示，Education在整体上对Balance的影响不大，各类别相差并不显著，但值得注意的是，受教育年限为9年的数据偏高而受教育年限为13年的数据明显偏低

```python
sns.boxplot(x='Education', y='Balance', data=credit)
```

如下的代码框绘制出Cards的箱线图
